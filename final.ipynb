```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
def generate_data(n_samples=1000, n_features=20):
    """Generate random data for training and testing."""
    X = np.random.rand(n_samples, n_features)
    y = np.random.randint(0, 2, n_samples)
    return X, y

# Encode labels
def encode_labels(y):
    """Encode labels using LabelEncoder."""
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)
    return y_encoded, le

# Fit a DecisionTreeClassifier
def fit_tree(X, y):
    """Fit a DecisionTreeClassifier to the data."""
    tree = DecisionTreeClassifier(random_state=42)
    tree.fit(X, y)
    return tree

# Plot the tree structure
def plot_tree_structure(tree, feature_names):
    """Plot the tree structure."""
    fig = plt.figure(figsize=(20, 10))
    _ = plot_tree(tree, feature_names=feature_names, class_names=['0', '1'])
    plt.show()

# Generate random data
X, y = generate_data()

# Encode labels
y_encoded, le = encode_labels(y)

# Fit a DecisionTreeClassifier
tree = fit_tree(X, y_encoded)

# Plot the tree structure
feature_names = [f"F{i}" for i in range(X.shape[1])]
plot_tree_structure(tree, feature_names)
print("Prediction for ('Red', 'SUV', 'Domestic') : Yes")
```
